```python
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq

csv_path = "data/2019-Nov.csv"
parquet_path = "data/2019-Nov.parquet"
chunk_size = 100_000  # adjust based on your available RAM

# Open a ParquetWriter outside the chunk loop
writer = None

for i, chunk in enumerate(pd.read_csv(csv_path, chunksize=chunk_size, parse_dates=['event_time'])):
    # Ensure 'event_time' is a datetime
    chunk['event_time'] = pd.to_datetime(chunk['event_time'], errors='coerce')

    table = pa.Table.from_pandas(chunk)
    if writer is None:
        writer = pq.ParquetWriter(parquet_path, table.schema)
    writer.write_table(table)
    print(f"Processed chunk {i+1}")

if writer:
    writer.close()
    print(f"Successfully written to {parquet_path}")
```
